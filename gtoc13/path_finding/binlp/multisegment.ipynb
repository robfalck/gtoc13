```python
pidxs_params_1 = IndexParams(
    bodies_ID=k_body_1,
    n_timesteps=num_1,
    seq_length=7,
    flyby_limit=1,
    gt_planets=7,
    dv_limit=10.0,  # km/s
    first_arcs=[(10, 9, 8, 7), (10, 9, 8, 7)],
)
solv_params = SolverParams(
    solver_name="scip",  # AMPL-format solvers
    soln_gap=0.1,
)
segment_1, flyby_history, model_1 = run_segment_problem(
    pidxs_params_1, discrete_data_1, solv_params, dv_table_1, disallowed=1000
)
```
